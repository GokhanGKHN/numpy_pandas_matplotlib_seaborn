```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
veriSeti=pd.read_csv("./insurance.csv") # dosya ekleme
```

```python
veriSeti
```

```python
veriSeti = veriSeti.rename(
    columns={"age" : "yas",
             "sex" : "cinsiyet",
             "bmi" : "vki",
             "children" : "cocukSayisi",
             "smoker" : "sıgaraDurum",
             "charges" : "odemeMiktarı",
             "region" : "bolge"
            }) #yeniden adlandırma
```

```python
veriSeti
```

## Veri Tipi Kontrolü

```python
veriSeti.dtypes
```

## Niteliklerin Tükçeleştirilmesi

```python
veriSeti["cinsiyet"] = veriSeti["cinsiyet"].replace(["male","female"],["erkek","kadın"])
veriSeti["sıgaraDurum"] = veriSeti["sıgaraDurum"].replace(["no","yes"],["hayır","evet"])
veriSeti["bolge"] = veriSeti["bolge"].replace(["southeast","northeast","southwest","northwest"],["guneydogu","kuzeybati","güneybati","kuzeydoğu"])
```

## Bazı durumlarda aynı grafik alanında birden fazla grafiğin görüntülenmesi istenebilir

```python
#plot 1: ilk 20 müşterinin vki
x=np.arange(1,21)
y1=veriSeti.iloc[0:20,2]

plt.subplot(2,1,1) #2 satır 1 sütun ve grafik 
plt.plot(x,y1)
plt.xticks(x)
plt.title("ilk 20 müşteri")

y2 = veriSeti.iloc[20:40,2]
plt.subplot(2,1,2)
plt.plot(x,y2)
plt.title("İlk 20 ile 40 Müşteri")
plt.xticks(x)

plt.suptitle("vki karşılaştrı")
plt.tight_layout(pad=1)
```

```python
#plot 1: ilk 20 müşterinin vki
x=np.arange(1,21)
y1=veriSeti.iloc[0:20,2]

plt.subplot(1,2,1) #2 satır 1 sütun ve grafik 
plt.plot(x,y1)
plt.xticks(x)
plt.title("ilk 20 müşteri")

y2 = veriSeti.iloc[20:40,2]
plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title("İlk 20 ile 40 Müşteri")
plt.xticks(x)

plt.suptitle("vki karşılaştrı")
plt.tight_layout()
```

## Sütun Grafiği

```python
ozet= veriSeti.groupby("sıgaraDurum")["odemeMiktarı"].mean()

def etiketEkle(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i], ha="center")

plt.bar(x=ozet.index, height=ozet.values, color="r",width=0.95)
plt.xlabel("Sıgra içme durumu")
plt.ylabel("Odeme Miktar")

etiketEkle(ozet.index,ozet.values.round(2))
```

```python
ozeth= veriSeti.groupby("sıgaraDurum")["odemeMiktarı"].mean()
plt.barh(y=ozeth.index, width=ozeth.values, color="r",height=0.2)
plt.xlabel("Sıgra içme durumu")
plt.ylabel("Odeme Miktar")
```

## Pasta Grafiği

```python
ozetpi= veriSeti.groupby("bolge")["odemeMiktarı"].sum()
etiketler =ozetpi.index
degerler =ozetpi.values
renkler=sns.color_palette("viridis",4)
secım = (0,0,0.2,0)
plt.pie(degerler, explode=secım, labels=etiketler,autopct="%%%4.1f",shadow=True
       ,startangle=360,colors=renkler)

```

```python
veriSeti["vki"].describe()
```

```python
bp= plt.boxplot(veriSeti["vki"],
                vert=True,
                showmeans=True,
                meanline =True,
                labels=("x"),
                patch_artist=True,
                medianprops ={'linewidth' : 2,
                              'color' : 'purple'
                             },
                meanprops ={'linewidth' :2, 
                            'color' :'red'    
                           },
                notch=True)
plt.ylabel("Sigortaya Odenen Miktar")
plt.legend([bp['medians'][0],bp['means'][0]],['median','mean'])
```
